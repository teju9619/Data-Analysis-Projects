```python
most_common_country_per_director = data.groupby('director')['country'].agg(lambda x: x.mode().iat[0] if not x.mode().empty else 'Unknown').to_dict()

def impute_country_based_on_director(row): 
    if row['country'] == 'Unknown':
        return most_common_country_per_director.get(row['director'], 'Unknown') 
    return row['country']

data['country'] = data.apply(impute_country_based_on_director, axis=1) 
data
```
